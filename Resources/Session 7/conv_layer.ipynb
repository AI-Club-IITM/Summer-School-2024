#Imports

```python
import matplotlib.pyplot as plt #For plotting functionality
```

```python
import torch           #To incorporate Tensors and their functionality
import torch.nn as nn  #Provides all the functionality required to develop Neural Networks
```

#Example Image

```python
#Setting the seed
torch.manual_seed(20)
```

```python
#Treat this random Tensor of shape (3, 256, 256) as our example Image
image = torch.randint(0, 256, (3, 256, 256))

#Converting it to float data type
image = (image).float()
```

```python
#size of the image
print (image.shape)

#Printing the image's pixels
print (image)
```

```python
#Displaying the random image
numpy_image = image.numpy()
numpy_image = numpy_image.transpose(1, 2, 0)
plt.imshow(numpy_image)
plt.show()
```

#How a Convolutional Layer works

```python
#Initialising a Convolutional layer
conv_layer = nn.Conv2d (in_channels = 3, out_channels = 2, kernel_size = 3, stride = 1, padding = 0)
```

```python
#Image before passing it through the layer
print (image.shape)
print (image)
```

```python
#Passing it through the layer
output = conv_layer (image)
```

```python
#Printing output
print (output.shape)
print (output)
#output shape verifies our formula for n_out
```

#Parameters of the Convolutional Layer

```python
#Kernels involved (Here they are referred to as weights)
print (conv_layer.weight.shape)
print (conv_layer.weight)
```

```python
#Biases of the layer
print (conv_layer.bias.shape)
print (conv_layer.bias)
```

#Verifying output channel calculation

```python
print (output)
```

```python
first_element = output[0][0][0]       #First element of the first output channel
print (first_element)
```

```python
kernel_1_1 = conv_layer.weight[0][0]   #First Kernel for producing the first output channel
print (kernel_1_1)
```

```python
kernel_1_2 = conv_layer.weight[0][1]   #Second Kernel for producing the first output channel
print (kernel_1_2)
```

```python
kernel_1_3 = conv_layer.weight[0][2]   #Third Kernel for producing the first output channel
print (kernel_1_3)
```

```python
#Image channels
image_1 = image[0]
image_2 = image[1]
image_3 = image[2]
print (image_1)
```

```python
#elements of first channel which the Kernel covers
image_1_1 = image_1[:3, :3]
print (image_1_1)
```

```python
#elements of first channel which the Kernel covers
image_2_1 = image_2[:3, :3]
print (image_2_1)
```

```python
#elements of first channel which the Kernel covers
image_3_1 = image_3[:3, :3]
print (image_3_1)
```

```python
# performing dot - products

dot_1 = (image_1_1 * kernel_1_1)
dot_2 = (image_2_1 * kernel_1_2)
dot_3 = (image_3_1 * kernel_1_3)

print (dot_1)
print (dot_2)
print (dot_3)
```

```python
dot_1 = torch.sum (dot_1)
dot_2 = torch.sum (dot_2)
dot_3 = torch.sum (dot_3)

print (dot_1)
print (dot_2)
print (dot_3)
```

```python
bias_1 = conv_layer.bias[0]
print (bias_1)
```

```python
element_1 = dot_1 + dot_2 + dot_3 + bias_1
print (element_1)
print (first_element)           #Both results match
```
