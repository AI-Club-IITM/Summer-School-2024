```python
#Import required libaries
import numpy as np
import pandas as pd
```

# Data preprocessing

```python
#Reading the melbourne housing dataset csv file
dataset = pd.read_csv("Melbourne_housing_FULL.csv")
dataset.head()
```

```python
#Printing out the number of unique entries of each column
dataset.nunique()
```

```python
#Selecting our feature columns and target(Price)
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG', 'Regionname','Propertycount','Distance','CouncilArea',
               'Bedroom2','Bathroom','Car','Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
#Total NaN entries in each column
dataset.isna().sum()
```

```python
#Fill NaN values of a some columns with 0
cols_to_fill_zero = ['Bedroom2','Bathroom','Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
#Fill NaN values with mean
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset['Distance'] = dataset['Distance'].fillna(dataset.Distance.mean())

#Fill NaN values with median
dataset['Propertycount'] = dataset['Propertycount'].fillna(dataset.Propertycount.median())


dataset.isna().sum()
```

```python
#Drop the remaining rows with NaN values
dataset.dropna(inplace = True)
dataset.isna().sum()
```

```python
#Dummy encoding of non numeric data
dataset = pd.get_dummies(dataset, drop_first = True)
dataset.head()
```

# Performing Linear, ridge and lasso regression and comparing results

```python
X = dataset.drop('Price', axis =1) #Features
y = dataset['Price']  #Target
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y ,test_size = 0.3, random_state = 2)
```

```python
#Performing Linear Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

```python
#Performing Lasso Regression
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
#Performing Ridge regression
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha = 50, max_iter = 100, tol = 0.1)
ridge_reg.fit(X_train, y_train)
```

```python
ridge_reg.score(X_test, y_test)
```

```python
ridge_reg.score(X_train, y_train)
```
