Nested Loop:

Loops within loops

```python
x = 4
for i in range(1, x + 1):
  for j in range(1, i + 1):
    print(j, end = "")
  print()
```

While loop:

```python
number = int(input('Enter a number: '))

total = 0

# iterate until the user enters 0
while number != 0:
    total += number
    number = int(input('Enter a number: '))

print('The sum is', total)
```

```python
# example class

class Book:          # class is a keyword
  # class variable: shared by all instances of a class
  Type = "Fiction"
  Name = "Harry Potter"
  price = "300"

  # member function/ method
  def reading(self):
    print('read')

### INSTIAITING AN OBJECT
c1 = Book()        # creating an instance of book
c1.reading()        # how to call a method under a class
```

image.png

```python
class Vehicle:

    # class variable: shared by all instances of a class
    vehicle_type = 'car'

    # init method or constructor
    def __init__(self, company1, color1):

        # instance variables

        self.company = company1             #here company is the attribute and company1 is the value
        self.color = color1

    # method to display details
    def printing(self):
        print('Vehicle type: ', self.vehicle_type)
        print('company: ', self.company)
        print('color: ', self.color)



C1 = Vehicle("Ferrari", "red")
C2 = Vehicle("Mercedes", "gold")

# displays details of C1
print('C1 details: ', end = '')
C1.printing()

# accessing using class names, attributes can be accessed viw the . operator
print("\naccessing class variable using class name")
print(C1.color)
print(C2.company)
```

INHERITANCE: capability of one class to derive or inherit the properties from another class.

image.png

```python
class company:
  def __init__(self):
    self.ceo = "mr X"
    self.company_name = "ABC"

  def show(self):
    print(f"Company : {self.company_name}")
    print(f"CEO : {self.ceo}")

class Salesemployee(company):
   def __init__(self,name):
     company.__init__(self)
     self.name = name
     self.salary = "20000"

   def printsalary(self):
     print(f"Salary:{self.salary}")

class Serviceemployee(company):
   def __init__(self,name,branch_name):
     company.__init__(self)
     self.branch_name  = branch_name

   def branch(self):

     print(f"Branch:{self.branch_name}")

emp1 = Salesemployee("mr A")
print(emp1.show())
print(emp1.printsalary())

print("\n")

emp2 = Serviceemployee("mr B","Delhi")
print(emp2.show())
print(emp2.branch())
```
